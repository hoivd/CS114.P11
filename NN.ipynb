```python
i = 1
print(f'Fold processing: {i}')
df_train = pd.read_pickle(f'/kaggle/input/extract-feature-train-5-fold/train_features_labels_fold_{i}.pkl')

X_train = np.array(df_train['feature'].tolist(), dtype=np.float32)
y_train = np.array(df_train['label'], dtype=np.int8)
                  

del df_train

df_test = pd.read_pickle(f'/kaggle/input/extract-feature-val-5-fold/val_features_labels_fold_{i}.pkl')
X_test = np.array(df_test['feature'].tolist(), dtype=np.float32)
y_test = np.array(df_test['label'], dtype=np.int8)


model = models.Sequential([
    Input(shape=(47040,)),
# layers.Flatten(input_shape=X_train[0].shape),
    layers.Dense(512),
    LeakyReLU(alpha=0.01),
    layers.BatchNormalization(),

    layers.Dense(256),
    LeakyReLU(alpha=0.01),
    layers.BatchNormalization(),

    layers.Dropout(0.2),

    layers.Dense(9, activation='softmax') # 9 classes 
])
optimizer = Adam(learning_rate=0.0025)

model.compile(optimizer=optimizer,
          loss='sparse_categorical_crossentropy',
          metrics=['accuracy']
          )

model.fit(X_train, y_train,batch_size=6 ,epochs=50, verbose=1)

# Đánh giá mô hình trên tập test
# model.save(f"/kaggle/working/models/model_fold_{i}.h5")

score = model.evaluate(X_test, y_test, verbose=0)
print(score)
del df_test

gc.collect()
tf.keras.backend.clear_session()
tf.compat.v1.reset_default_graph()
```
